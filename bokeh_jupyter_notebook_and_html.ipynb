# ColumnDataSource

```python
import bokeh
import bokeh.plotting
p = bokeh.plotting.figure(plot_width=200, plot_height=200,x_range=(0,4),y_range=(0,4))

data_source = bokeh.models.ColumnDataSource(dict(image=[[[1,3.],[2,4]],
                                                        [[1,3.,2],[1,2,4],[3,3,2]],
                                                       ],
                                                 x=[0,2],
                                                 dw=[2,2],
                                                 y=[0,2],
                                                 dh=[2,2]))

p.image(image="image", x="x", dw="dw", y="y", dh="dh",source = data_source)
bokeh.io.show(p)
```

# layouts and widgets and CustomJS

```python
import bokeh
import bokeh.plotting
bokeh.io.output_notebook()
p = bokeh.plotting.figure(plot_width=200, plot_height=200,x_range=(0,4),y_range=(0,4))

data_source = bokeh.models.ColumnDataSource(dict(image=[[[1,3.],[2,4]],
                                                        [[1,3.,2],[1,2,4],[3,3,2]],
                                                       ],
                                                 x=[0,2],
                                                 dw=[2,2],
                                                 y=[0,2],
                                                 dh=[2,2]))

p.image(image="image", x="x", dw="dw", y="y", dh="dh",source = data_source)

callback_01 = bokeh.models.CustomJS(args=dict(source=data_source),code="""
    // Java Script code
    console.log("Button pressed");
    
    var image_index=1;
    var d = source.data['image'][image_index];
    console.log(d);
    var rows = d.length;
    for (var row = 0; row < rows; row++){
        var cols = d[row].length;
        for (var col = 0; col < cols; col++) {
            d[row][col] = Math.random();
        }
    }
    source.change.emit();
""")

button = bokeh.models.Button(label="Modify Data",callback=callback_01)
layout = bokeh.layouts.row(p,button)
bokeh.io.show(layout)
```

# standalone html

```python
bokeh.io.output_file("example3.html")
bokeh.io.save(layout)
```
